# **Compare the performance (time taken) of C, C++, Java, R, and Python programs**

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

# Read the data from the CSV file
df = pd.read_csv('performance_comparison.csv')

# Extract the data into lists
x = df['File Size (MB)'].tolist()
y1 = df['C (sec)'].tolist()
y2 = df['C++ (sec)'].tolist()
y3 = df['Java (sec)'].tolist()
y4 = df['R (sec)'].tolist()
y5 = df['Python (sec)'].tolist()

# Create the lineplot
pics = sns.lineplot(x = x, y = y1, label='C', linewidth=2)
sns.lineplot(x = x, y = y2, label='C++', linewidth=1)
sns.lineplot(x = x, y = y3, label='Java', linewidth=3)
sns.lineplot(x = x, y = y4, label='R', linewidth=2)
sns.lineplot(x = x, y = y5, label='Python', linewidth=2)
pics.set_xlabel('File size(GB)')
pics.set_xlabel('Time Taken(sec)')
pics.set_title('Performance of different programming languages')
```
